```python
doc1 = 'Data is the oil of the digital economy'
doc2 = 'Data is a new oil'
data = [doc1,doc2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer()
vector_matrix = CV.fit_transform(data)
vector_matrix
```

```python
token = CV.get_feature_names_out()
```

```python
token
```

```python
vector_matrix.toarray()
```

```python
def create_dataframe(matrix,tokens):
  doc_names = [f'doc_{i+1}' for i, _ in enumerate(matrix)]
  df = pd.DataFrame(data = matrix,index = doc_names, columns = tokens)
  return df
```

```python
create_dataframe(vector_matrix.toarray(),token)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = cosine_similarity(vector_matrix)
create_dataframe(cosine_similarity_matrix,['doc 1', 'doc_2'])
```

# ***Project -Movie Recommendation System***

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits,on = 'title')
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
```

```python
def convert(text):
  L = []
  for i in ast.literal_eval(text):
    L.append(i['name'])
  return L
```

```python
movies.dropna(inplace = True)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast']=movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
  L = []
  for i in ast.literal_eval(text):
    if(i['job'] == 'Director'):
      L.append(i['name'])
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
def collapse(L):
  L1 = []
  for i in L:
    L1.append(i.replace(' ',''))
  return L1
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns = ['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vector)
```

```python
sim
```

```python
def recommend(movie):
    indices = new[new['title'] == movie].index
    if len(indices) == 0:
        print(f"Error: Movie '{movie}' not found in the dataset.")
        return  # or raise an exception, or return a default value
    index = indices[0]
    distances = sorted(list(enumerate(sim[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```

```python
recommend('The Lego Movie')
```
